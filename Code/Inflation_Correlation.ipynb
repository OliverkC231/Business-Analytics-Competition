```python
import pandas as pd

# Assuming debt_data is already loaded
# Ensure 'Country' and 'ccode' are columns in your DataFrame
# You might want to move them to the front of the DataFrame if needed
debt_data = debt_data[['Country', 'CCode'] + [str(year) for year in range(2010, 2024)]]  # Ensuring correct column order

# Convert the debt columns (2010-2023) to numeric, handling errors
for col in debt_data.columns[2:]:  # Start from the 3rd column, which are years 2010-2023
    debt_data[col] = pd.to_numeric(debt_data[col], errors='coerce')  # Convert to numeric, replace non-numeric with NaN

# Calculate the percentage change for YoY (Year-over-Year) change
debt_yoy = debt_data[[str(year) for year in range(2011, 2024)]].pct_change(axis='columns')  # pct_change() calculates percentage change year-over-year

# Add back the 'Country' and 'ccode' columns to the result
debt_yoy = pd.concat([debt_data[['Country', 'CCode']], debt_yoy], axis=1)

# Print the first few rows to verify
debt_yoy.head()
```

```python
from scipy.stats import pearsonr
import pandas as pd

# Set 'Country' as the index for both dataframes
debt_yoy.set_index('Country', inplace=True)
inf_data.set_index('Country', inplace=True)

# Filter the columns for 2010-2023 (assuming these are the year columns)
debt_yoy = debt_yoy.loc[:, '2012':'2023']  # Keep only 2012-2023 columns
inf_data = inf_data.loc[:, '2012':'2023']  # Keep only 2012-2023 columns

# Transpose the dataframes so that years become the index
debt_yoy = debt_yoy.T
inf_data = inf_data.T

# Merge datasets by year
merged_data = pd.merge(debt_yoy, inf_data, left_index=True, right_index=True, suffixes=('_debt', '_inflation'))

# Drop rows with NaN values
merged_data = merged_data.dropna()

# Calculate Pearson correlation by year
correlations_by_year = {}

for year in merged_data.index:
    year_data = merged_data.loc[year]  # Data for a single year
    year_corr, _ = pearsonr(year_data.filter(like='_debt'), year_data.filter(like='_inflation'))  # Correlation between debt and inflation for the year
    correlations_by_year[year] = year_corr

# Output the results
print("Correlation coefficients for each year between public sector debt and inflation:")
for year, corr in correlations_by_year.items():
    print(f"{year}: {corr}")
```

```python
# Calculate the average correlation coefficient
average_corr = sum(correlations_by_year.values()) / len(correlations_by_year) if correlations_by_year else None

# Output the average correlation
print(f"\nThe average Pearson correlation coefficient between YoY public sector debt and inflation across all countries is: {average_corr}")
```
