```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/Week 7/World Bank Real GDP Data.csv')
```

```python
df_cleaned = df.iloc[2:].reset_index(drop=True)
df_cleaned.columns = df_cleaned.iloc[0]
df_cleaned = df_cleaned[1:].reset_index(drop=True)

# Rename first two columns for clarity
df_cleaned = df_cleaned.rename(columns={'Country Name': 'Country', 'Country Code': 'Country Code'})

# Convert column names to string for consistency and remove decimal points
df_cleaned.columns = df_cleaned.columns.map(lambda x: str(int(float(x))) if str(x).replace('.0', '').isdigit() else x)

# List of countries to filter
filter_countries = [
    'Canada', 'Mexico', 'United States', 'Austria', 'Belgium', 'France', 'Germany', 'Ireland',
    'Luxembourg', 'Netherlands', 'Switzerland', 'United Kingdom', 'Denmark', 'Estonia', 'Finland',
    'Iceland', 'Latvia', 'Lithuania', 'Norway', 'Sweden', 'Greece', 'Italy', 'Portugal', 'Spain',
    'Turkey', 'Czech Republic', 'Hungary', 'Poland', 'Slovak Republic', 'Slovenia', 'Australia',
    'Japan', 'South Korea', 'New Zealand', 'Chile', 'Colombia', 'Costa Rica', 'Turkiye',
    'Korea, Rep.', 'Czechia'
]

# Filter the dataframe
df_filtered = df_cleaned[df_cleaned['Country'].isin(filter_countries)]

# Keep only data from 2010 to 2023
columns_to_keep = ['Country'] + [str(year) for year in range(2010, 2024)]
df_filtered_years = df_filtered[columns_to_keep]

# Save the filtered dataset
df_filtered_years.to_csv("Cleaned_GDP_Data_2010_2023.csv", index=False)
```

```python
pd.read_csv('/content/drive/MyDrive/Machine Learning/Week 7/National GDP in USD (1).csv')
```

```python
df6 = pd.read_csv('/content/drive/MyDrive/Machine Learning/Week 7/National GDP in USD (1).csv')
```

```python
# Rename for consistency
df6.columns = df6.columns.astype(str)
df6 = df6.rename(columns={'Country Name': 'Country', 'Country Code': 'Country Code'})

# Filter countries
filter_countries = [
    'Canada', 'Mexico', 'United States', 'Austria', 'Belgium', 'France', 'Germany', 'Ireland',
    'Luxembourg', 'Netherlands', 'Switzerland', 'United Kingdom', 'Denmark', 'Estonia', 'Finland',
    'Iceland', 'Latvia', 'Lithuania', 'Norway', 'Sweden', 'Greece', 'Italy', 'Portugal', 'Spain',
    'Turkey', 'Czech Republic', 'Hungary', 'Poland', 'Slovak Republic', 'Slovenia', 'Australia',
    'Japan', 'South Korea', 'New Zealand', 'Chile', 'Colombia', 'Costa Rica', 'Turkiye',
    'Korea, Rep.', 'Czechia'
]
df6_filtered = df6[df6['Country'].isin(filter_countries)]

# Save Country Code info separately before modifying the structure
country_codes = df6_filtered[['Country', 'Country Code']].copy()

# Keep only 2009-2023
columns_to_keep = ['Country'] + [str(year) for year in range(2009, 2024)]
df6_filtered = df6_filtered[columns_to_keep]

# Set index and calculate YoY
df6_filtered.set_index('Country', inplace=True)
df6_yoy = df6_filtered.div(df6_filtered.shift(axis=1)) * 100
df6_yoy = df6_yoy.loc[:, '2010':]  # Keep only 2010-2023
df6_yoy.reset_index(inplace=True)

# Merge Country Code back into the YoY DataFrame
df6_yoy = pd.merge(df6_yoy, country_codes, on='Country')

# Reorder columns to have Country, Country Code, then years
cols = ['Country', 'Country Code'] + [str(year) for year in range(2010, 2024)]
df6_yoy = df6_yoy[cols]

# Save output
df6_yoy.to_csv("National_GDP_YoY_2010_2023_filtered2.csv", index=False)
```
