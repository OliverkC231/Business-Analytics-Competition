```python
import pandas as pd

# Load datasets
debt_df = pd.read_csv('/content/drive/MyDrive/Saint Mary 2024-2025/Data 496/Dataset to make debt to GDP/Public_Sector_Debt.csv')
gdp_df = pd.read_csv('/content/drive/MyDrive/Saint Mary 2024-2025/Data 496/Dataset to make debt to GDP/GDP_Per_Capita_Data_2010_2023 - Cleaned_GDP_Per_Capita_Data_2010_2023.csv')

# Identify year columns (2010 to 2023)
year_columns = [str(year) for year in range(2010, 2024)]

# Merge datasets on 'CCode' (keeping all columns)
merged_df = pd.merge(debt_df, gdp_df, on=['CCode'], how='inner', suffixes=('_Debt', '_GDP'))

# Compute Debt-to-GDP ratio for each year
for year in year_columns:
    merged_df[year] = (merged_df[year + '_Debt'] / merged_df[year + '_GDP'])

# Keep only relevant columns: Country info and ratio columns
final_columns = ['Country Name', 'CCode'] + [year for year in year_columns]
debt_to_gdp_df = merged_df[final_columns]

# Display results
print(debt_to_gdp_df.head())

# Save results to CSV
debt_to_gdp_df.to_csv('/content/drive/MyDrive/Saint Mary 2024-2025/Data 496/Dataset to make debt to GDP/Debt_to_GDP_Ratio.csv', index=False)
```
