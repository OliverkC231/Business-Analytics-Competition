```python
import pandas as pd
```

```python
urbanpop = pd.read_csv("/content/Urban Population (% of total population).csv")
urbanpop.head()
```

```python
countries_to_keep = [
    "Canada", "Mexico", "United States", "Austria", "Belgium", "France", "Germany", "Ireland", "Luxembourg", "Netherlands",
    "Switzerland", "United Kingdom", "Denmark", "Estonia", "Finland", "Iceland", "Latvia", "Lithuania", "Norway", "Sweden",
    "Greece", "Italy", "Portugal", "Spain", "Turkey", "Czechia", "Hungary", "Poland", "Slovak Republic", "Slovenia",
    "Australia", "Japan", "South Korea", "New Zealand", "Chile", "Colombia", "Costa Rica"
]

urbanpop_filtered = urbanpop[urbanpop["Country"].isin(countries_to_keep)]
urbanpop_filtered.head()
```

```python
import pandas as pd

# Filter to keep only rows where Year is between 2009 and 2023
urbanpop_filtered= urbanpop_filtered[(urbanpop_filtered['Year'] >= 2009) & (urbanpop_filtered['Year'] <= 2023)]
```

```python
urbanpop_filtered.head()
```

```python
# Pivot the data so years become columns
urbanpop_filtered = urbanpop_filtered.pivot(index='Country', columns='Year', values='Urban population (% of total population)').reset_index()
urbanpop_filtered.head()
```

```python
print(len(urbanpop_filtered))
```

```python
urbanpop_filtered.to_csv('filtered_urban_population(% of total polpulation).csv', index=False)
```

```python
# If Country is a column, set it as the index first
urbanpop_filtered.set_index('Country', inplace=True)

# Calculate YoY change
urbanpop_yoy = urbanpop_filtered.pct_change(axis=1) * 100  # Multiply by 100 to express as percentage

# Keep only years 2010 to 2023
urbanpop_yoy = urbanpop_yoy.loc[:, 2010:2023]

# Reset index if needed
urbanpop_yoy = urbanpop_yoy.reset_index()

# Optional: round to 2 decimal places
urbanpop_yoy = urbanpop_yoy.round(2)
```

```python
urbanpop_yoy.head()
```

```python
urbanpop_yoy.to_csv('filtered_urban_population_YOY(% of total polpulation).csv', index=False)
```

```python
corruption = pd.read_csv("/content/Corruption Index Combined Data.csv")
corruption.head()
```

```python
corruption = corruption[corruption["Country"].isin(countries_to_keep)]
corruption.head()
```

```python
print(len(corruption))
```

```python
# Start by extracting all column names
cols = corruption.columns.tolist()

# Keep the 'Country' column first
country_col = ['Country']

# Extract and sort the CPI score columns by year (2009 to 2023)
cpi_cols = sorted([col for col in cols if 'CPI score' in col or 'CPI Score' in col],
                  key=lambda x: int(x.split()[-1]))

# Combine them
ordered_cols = country_col + cpi_cols

# Reorder the DataFrame
corruption = corruption[ordered_cols]
```

```python
corruption.head()
```

```python
# Rename columns: keep 'Country', and extract year from other columns
corruption.columns = ['Country' if col == 'Country' else col.split()[-1] for col in corruption.columns]
```

```python
corruption.head()
```

```python
corruption.to_csv('filtered_corruption_score(0-100)).csv', index=False)
```

```python
# If Country is a column, set it as the index first
corruption.set_index('Country', inplace=True)

# Calculate YoY change
corruption_yoy = corruption.pct_change(axis=1) * 100  # Multiply by 100 to express as percentage

# Keep only years 2010 to 2023
corruption_yoy = corruption_yoy.loc[:, '2010':'2023']

# Reset index if needed
corruption_yoy = corruption_yoy.reset_index()

# Optional: round to 2 decimal places
corruption_yoy = corruption_yoy.round(2)
```

```python
corruption_yoy.head()
```

```python
corruption_yoy.to_csv('filtered_corruption_score_YOY(0-100)).csv', index=False)
```
