```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV files
df1 = pd.read_csv('/content/Public_Sector_Debt.csv')
df2 = pd.read_csv('/content/clean_import_of_goods_and_services_as_a_percentage_of_GDP.csv')
```

```python
# Inspect column names
print("Columns in df1:", df1.columns.tolist())
print("Columns in df2:", df2.columns.tolist())

# Merge DataFrames on 'Country Name'
merged_df = pd.merge(df1, df2, on='Country Name', suffixes=('_df1', '_df2'))

# Inspect merged DataFrame columns
print("Columns in merged_df:", merged_df.columns.tolist())

# List of years to calculate correlation for
years = [str(year) for year in range(2010, 2024)]  # From 2010 to 2023

# Calculate correlation for each year
correlation_results = {}
for year in years:
    col_df1 = f"{year}_df1"  # Column name for df1 (e.g., '2010_df1')
    col_df2 = f"{year}_df2"  # Column name for df2 (e.g., '2010_df2')

    # Calculate correlation
    correlation = merged_df[col_df1].corr(merged_df[col_df2])
    correlation_results[year] = correlation

# Print correlation results
print("Correlation between yearly values:")
for year, corr in correlation_results.items():
    print(f"{year}: {corr}")
```

```python
# Inspect column names
print("Columns in df1:", df1.columns.tolist())
print("Columns in df2:", df2.columns.tolist())

# Merge DataFrames on 'Country Name'
merged_df = pd.merge(df1, df2, on='Country Name', suffixes=('_df1', '_df2'))

# Inspect column names in the merged DataFrame
print("Columns in merged_df:", merged_df.columns.tolist())

# List of common years (2010 to 2021)
common_years = [str(year) for year in range(2010, 2024)]  # 2010 to 2023

# Create a correlation matrix for exports (df1) and government expenditure (df2) for common years
correlation_matrix = merged_df[[f'{year}_df1' for year in common_years] +
                      [f'{year}_df2' for year in common_years]].corr()

# Filter the correlation matrix to show only exports vs. government expenditure for the same year
exports_vs_expenditure_corr = correlation_matrix.loc[[f'{year}_df1' for year in common_years],
                                                     [f'{year}_df2' for year in common_years]]

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    exports_vs_expenditure_corr,
    annot=True,            # Annotate cells with correlation values
    cmap='coolwarm',       # Color map (blue to red)
    vmin=-1,               # Minimum correlation value
    vmax=1,                # Maximum correlation value
    fmt=".2f",             # Format annotations to 2 decimal places
    linewidths=0.5,        # Add lines between cells
    xticklabels=common_years,  # Label x-axis with years
    yticklabels=common_years   # Label y-axis with years
)
plt.title('Correlation Heatmap: Public Sector Debt vs. Imports of Good and Services as % of GDP (2010-2023)')
plt.xlabel('Imports of Good and Services as % of GDP (Years)')
plt.ylabel('Public Sector Debt (Years)')
plt.show()
```
